```python
pln = spacy.load('pt_core_news_sm')
punct = string.punctuation
doc = nlp(corpus)
doc2 = nlp("Não terei medo. O medo mata a mente. O medo é a pequena morte que leva à aniquilação total. Enfrentarei meu medo. Permitirei que passe por cima e através de mim. E quando estiver passado, voltarei o olho interior para ver seu rastro. Onde o medo não estiver mais, nada haverá. Somente eu restarei")
doc3 = nlp('Projeto em tramitação na Câmara dos Deputados será tema de conferência on-line, nesta quinta-feira, dia 14, do professor Carlos Roberto Jamil Cury, da Universidade Federal de Minas Gerais, que abrirá as atividades deste ano da Cátedra Alfredo Bosi de Educação Básica da USP')
```

```python
def preproc(texto):
  texto = texto.lower()
  uwu = pln(texto)

  lista = []
  for token in uwu:
    lista.append(token.lemma_)

  lista = [palavra for palavra in lista if palavra not in stopwords or palavra not in punct]
  lista = ' '.join([str(elemento) for elemento in lista if not elemento.isdigit()])
  return lista
```

```python
teste = preproc(str(doc2))
teste
```

```python
df['texto'] = df['texto'].apply(preproc)

dfowo = [["este trabalho é agradável", {"ALEGRIA": True, "MEDO": False}],
                      ["este lugar continua assustador", {"ALEGRIA": False, "MEDO": True}]]
```

```python
dados = []

for texto, emocao in zip(df['texto'], df['emocao']):
  if emocao == 'alegria':
    dic = ({"ALEGRIA": True, "MEDO": False})
  else:
    dic = ({"ALEGRIA": False, "MEDO": True})

  dados.append([texto, dic.copy()])
```

```python
from spacy.training import Example
```

```python
modelo = spacy.blank('pt')
textcat = modelo.add_pipe('textcat')
textcat.add_label('ALEGRIA')
textcat.add_label('MEDO')
historico = []
```

```python
modelo.begin_training()
for epoca in range(200):
  random.shuffle(dados)
  losses = {}
  for batch in spacy.util.minibatch(dados, 30):
    textos = [modelo(texto) for texto, entities in batch]
    annot = [{'cats': entities} for texto, entities in batch]
    examples = [Example.from_dict(doc,annot) for doc, annot in zip(textos, annot)]
    modelo.update(examples, losses=losses)
  if epoca % 100 == 0:
    print(losses)
    historico.append(losses)
```

```python
modelo.to_disk('modelo')
```

```python
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
from nltk.corpus import stopwords
from string import punctuation
from nltk.probability import FreqDist
from collections import defaultdict
from heapq import nlargest

link = Request("https://tecnologia.ig.com.br/2023-09-20/apple-libera-recurso-transforma-iphone-maquininha-cartao.html", headers={"User-Agent": "Mozilla/5.0"})
pagina = urlopen(link).read().decode('utf-8', 'ignore')
```

```python
soup = BeautifulSoup(pagina, 'lxml')
texto = soup.find(id='noticia').text
```

```python
sentencas = sent_tokenize(texto)
palavras = word_tokenize(texto.lower())
stopwords = set(stopwords.words('portuguese') + list(punctuation) + list('``'))
```

```python
freq = FreqDist(palavras)
sentencas_importantes = defaultdict(int)

for i, sentenca in enumerate(sentencas):
    for palavra in word_tokenize(sentenca.lower()):
        if palavra in freq:
            sentencas_importantes[i] += freq[palavra]

sentencas_importantes
```

```python
idx_sents = nlargest(4, sentencas_importantes, sentencas_importantes.get)
for i in sorted(idx_sents):
    print(sentencas[i])
```
